```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
recipes = [
    {
        "name": "Fish Curry",
        "ingredients": ["fish", "potato", "onion", "spices","oil"],
        "description": "Slice onions. Fry the fishes and onions in oil. Add spices and water. Boil for 20 minutes."
    },
    {
        "name": "Chicken ChowMein",
        "ingredients": ["chicken", "chowmein ", "sauce", "garlic", "ginger", "onions", "oil"],
        "description": "Cook chowmein. Stir-fry garlic and ginger in oil. Add chicken and cook. Mix the chowmein and soy sauce."
    },
    {
        "name": "Pani Puri",
        "ingredients": ["puris", "boiled potatoes", "spiced water", "tamarind chutney", "chaat masala"],
        "description": "Fill puris with potatoes mix with chaat masala. Add tamarind chutney to it. Top with spiced water and serve."
    },
    {
        "name": "Chicken Biryani",
        "ingredients": ["chicken", "basmati rice", "onions", "tomatoes", "biryani masala", "oil"],
        "description": "Marinate chicken with biryani masala. Cook rice. Fry onions in oil. Add chicken, tomatoes. Cook on low heat until done."
    }
]
```

```python
recipe_texts = [" ".join(i["ingredients"]) for i in recipes]
```

```python
recipe_texts
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
recipe_embeddings = model.encode(recipe_texts, convert_to_tensor=True)
```

```python
recipe_embeddings.shape
```

```python
# Using FAISS index we can perform efficient similarity search.
# When we query the index with a new embedding, FAISS will compute the distances between the query and all stored embeddings and return the most similar ones.
index = faiss.IndexFlatL2(recipe_embeddings.shape[1])
index.add(np.array(recipe_embeddings.cpu()))
```

```python
index.ntotal
```

```python
def retrieve_recipes(ingredients, k=1):
    query = " ".join(ingredients)
    query_embedding = model.encode([query], convert_to_tensor=True)
    distances, indices = index.search(np.array(query_embedding.cpu()), k)
    retrieved_recipes = [recipes[i] for i in indices[0]]
    return retrieved_recipes
```

```python
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
# Add padding token to the tokenizer
 #(GPT-2 does not have a padding token by default, so we use the EOS token as a padding token.)
tokenizer.pad_token = tokenizer.eos_token
gpt2_model = GPT2LMHeadModel.from_pretrained('gpt2') #text generation
```

```python
def generate_response(ingredients):
    retrieved_recipes = retrieve_recipes(ingredients, k=2)
    context = " ".join([recipe["description"] for recipe in retrieved_recipes])
    input_text = "Ingredients: " + ", ".join(ingredients) + ". Here is a recipe: " + context

    # We use encode_plus to get input IDs and attention mask( text into token IDs)
    encoded_inputs = tokenizer.encode_plus(
        input_text,
        return_tensors='pt', # PyTorch tensors
        padding=True,   # padded to a uniform length
        truncation=True,
        max_length=100
    )
    input_ids = encoded_inputs['input_ids']
    attention_mask = encoded_inputs['attention_mask']

    outputs = gpt2_model.generate(
        input_ids,
        max_length=150,
        num_return_sequences=1,
        no_repeat_ngram_size=3,
        pad_token_id=tokenizer.eos_token_id,
        attention_mask=attention_mask
    )
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response
```

```python
ingredients = ["fish", "potato", "onion", "spices","oil"]
response = generate_response(ingredients)
```
